```python
import pandas as pd
import math
```

```python
data = pd.read_excel('Courses Database.xlsx')
```

```python
courses_taken = ["MATH210", "MATH110", "CHEM100"]
courses_desire = ["FS100B", "CHEM101"]


#determine the credits you already earned
credit_total = 0
for course in courses_taken:
    credit = int(data.loc[data['course_number'] == course, 'credits'])
    credit_total = credit_total + credit
print("Your have already taken", credit_total, "credits.\n")


#check prerequisite
course_you_can_take = []
for course in courses_desire:
    prerequisite = data.loc[data['course_number'] == course, 'prerequisite'].iloc[0]

    if not pd.isna(prerequisite):
        prerequisite_list = prerequisite.split(', ')
        
        if all(req in courses_taken for req in prerequisite_list):
            course_you_can_take.append(course)
        
        elif "or" in prerequisite: 
            A = []
            for pre in prerequisite_list:
                if "or" in pre:
                    options = [opt.strip() for opt in pre.strip('()').split(' or ')]
                    if not any(option in courses_taken for option in options):
                        print("You need to take one of", options, "before taking", course)
                        A.append("no")
                else:
                    if pre not in courses_taken:
                        print("You need to take", pre, "before taking", course)
                        A.append("no")
            if not "no" in A:
                course_you_can_take.append(course)
                
        else:
            for req in prerequisite_list:
                if req not in courses_taken:
                    print("You need to take", req, "before taking", course)
        
    else:
        course_you_can_take.append(course)

if course_you_can_take == []:
    print("You cannot take all of these courses")
else:
    print("You can take", course_you_can_take, "next semester.")


#determine the credits you will earn
credits_will = 0
for course in course_you_can_take:
    credit = int(data.loc[data['course_number'] == course, 'credits'])
    credits_will = credits_will + credit

print("Your will earn", credits_will+credit_total, "credits after next semester.\n")


#check AA-degree
credits_Arts=0
credits_LL=0
credits_SMC=0
credits_SS=0
num_cp = 0
num_fs=0
course_total = courses_taken + course_you_can_take
num_courses = 0

for course in course_total:
    division = data.loc[data['course_number'] == course, 'division']
    division_name = division.item()
    cpfs = data.loc[data['course_number'] == course, 'CPFS']
    cpfs_if = cpfs.item()

    if division_name == "Science, Math, and Computing Division":
        credit1 = int(data.loc[data['course_number'] == course, 'credits'])
        credits_SMC = credits_SMC + credit1

    if division_name == "Arts Division":
        credit1 = int(data.loc[data['course_number'] == course, 'credits'])
        credits_Arts = credits_Arts + credit1

    if division_name == "Language and Literature Division":
        credit1 = int(data.loc[data['course_number'] == course, 'credits'])
        credits_LL = credits_LL + credit1

    if division_name == "Social Studies Division":
        credit1 = int(data.loc[data['course_number'] == course, 'credits'])
        credits_SS = credits_SS + credit1

    if cpfs_if == "cp":
        num_cp = num_cp + 1
    if cpfs_if == "fs":
        num_fs = num_fs + 1

    if credits_will+credit_total >= 60:
        num_courses = 1


if (credits_SMC>=6) and (credits_Arts>=6) and (credits_LL>=6) and (credits_SS>=6) and (num_cp>=1) and (num_fs==2) and (num_courses == 1):
    print("You will be able to get AA-degree after next semester.")
else:
    if credits_SMC < 6:
        print("You need to take",6-credits_SMC, "credits in Science, Math, and Computing Division")
    if credits_Arts < 6:
        print("You need to take",6-credits_Arts, "credits in Arts Division")
    if credits_LL < 6:
        print("You need to take",6-credits_LL, "credits in Language and Literature Division")
    if credits_SS < 6:
        print("You need to take",6-credits_SS, "credits in Social Studies Division")
    if num_cp==0:
        print("You need to take one CP course")
    if num_fs!=2:
        print("You need to take", 2-num_fs, "Seminar")
    
    if credits_will+credit_total < 60:
        print("You need to take",60-credits_will-credit_total, "credits in total")


#check time
begintime_set = []
for course in course_you_can_take:
    begin_time = data.loc[data['course_number'] == course, 'begin-time']
    begin_time_A = begin_time.item()
    if begin_time_A in begintime_set:
        print("You can't take two courses that are both beginning at", begin_time_A)
    else:
        begintime_set.append(begin_time_A)

endtime_set = []
for course in course_you_can_take:
    end_time = data.loc[data['course_number'] == course, 'end-time']
    end_time_A = end_time.item()
    if end_time_A in endtime_set:
        print("You can't take two courses that are both beginning at", end_time_A)
    else:
        endtime_set.append(begin_time_A)

```
